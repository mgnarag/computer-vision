## PART 6: Unsharp Masking in the Spatial Domain
Unsharp masking consists of blurring the original image and subtracting the blurred image from the original. We call the difference image as the ”mask”. The mask (or a fraction of it) is then added to the original.

Write code for the filter described above and apply unsharp masking to the building.tif image. Compare the filtered image with the original. What do you see that are not seen in the original ? Are they really there or are they artifacts of the process ? Explain



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('building.tif', cv2.IMREAD_GRAYSCALE)
blurred = cv2.GaussianBlur(image, (0, 0), sigmaX=3) #apply Gaussian blur with sigma = 3
mask = image - blurred #unsharp mask is obtained by subtracting  blurred image from original image

#Now, let's control the amount (fraction) of mask we want to add to the original:
def enhance(image,alpha):
    #we will use cv2.addWeighted function which performs a weighted sum of two input images
    #this allows us to control the contribution of the mask and the image to the final result
    enhanced_image = cv2.addWeighted(image, 1 + alpha, mask, -alpha, 0)
    return enhanced_image

#plotting
plt.figure(figsize = (15,5))
plt.subplot(131)
plt.imshow(image, cmap = 'gray')
plt.title('original image'), plt.xticks([]), plt.yticks([])

plt.subplot(132)
plt.imshow(blurred, cmap = 'gray')
plt.title('gaussian blue with sigma = 3'), plt.xticks([]), plt.yticks([])

plt.subplot(133),
plt.imshow(mask, cmap = 'gray')
plt.title('mask'), plt.xticks([]), plt.yticks([])

#plt.savefig('part6a.png')
plt.show()

plt.figure(figsize = (15,12))
plt.subplot(331)
plt.imshow(enhance(image,0.01), cmap = 'gray')
plt.title('enhanced image, alpha = 0.01'), plt.xticks([]), plt.yticks([])
plt.savefig('part6b.png')

plt.subplot(332)
plt.imshow(enhance(image,0.025), cmap = 'gray')
plt.title('enhanced image, alpha = 0.025'), plt.xticks([]), plt.yticks([])

plt.subplot(333)
plt.imshow(enhance(image,0.05), cmap = 'gray')
plt.title('enhanced image, alpha = 0.05'), plt.xticks([]), plt.yticks([])

plt.subplot(334)
plt.imshow(enhance(image,0.075), cmap = 'gray')
plt.title('enhanced image, alpha = 0.075'), plt.xticks([]), plt.yticks([])

plt.subplot(335)
plt.imshow(enhance(image,0.1), cmap = 'gray')
plt.title('enhanced image, alpha = 0.1'), plt.xticks([]), plt.yticks([])

plt.subplot(336)
plt.imshow(enhance(image,0.25), cmap = 'gray')
plt.title('enhanced image, alpha = 0.25'), plt.xticks([]), plt.yticks([])

plt.subplot(337)
plt.imshow(enhance(image,0.5), cmap = 'gray')
plt.title('enhanced image, alpha = 0.5'), plt.xticks([]), plt.yticks([])

plt.subplot(338)
plt.imshow(enhance(image,0.75), cmap = 'gray')
plt.title('enhanced image, alpha = 0.75'), plt.xticks([]), plt.yticks([])

plt.subplot(339)
plt.imshow(enhance(image,0.9), cmap = 'gray')
plt.title('enhanced image, alpha = 0.9'), plt.xticks([]), plt.yticks([])

#plt.savefig('part6b.png')
plt.show()
```

Unsharp filtering enhances edges and details of an image but it can also introduce artifacts. We need to carefully choose the parameters such as the amount of blurring and the fraction added back. If these parameters are chosen appropriately, the enhancement of edges and details can be real and visually pleasing, with minimal artifacts as can be seen in lower alphas = 0.01, 0.025, 0.05. If we go higher than these alpha values, the artifacts can become more pronounced. We can see these in alpha > 0.1
