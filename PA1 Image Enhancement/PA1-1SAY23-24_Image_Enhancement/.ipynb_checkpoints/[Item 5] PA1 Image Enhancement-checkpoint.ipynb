## (a) dental.gif

First, I will introduce the following openCV packages:

`cv2.inRange(src, lower_bound, upper_bound)`
-> creates a binary mask that isolates a specified range (lower_ and upper_bound) of pixel values in an input image.

`cv2.bitwise_and(src1, src2, mask = None)` 
-> performs bitwise AND operations on two images. It takes in src1 and src2 as the first and second image. Then the mask, if provided, the operation is only performed on the pixels where the corresponding pixel in the mask is non-zero. 


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


#First, i want to create a function that has threshold values as parameters
def gray_level_slicing(img, low_thresh, high_thresh):
    #we create a binary mask first to identify the pixels within the specified threshold range:
    mask = cv2.inRange(image, low_thresh, high_thresh) 
    #next, we create the sliced image with only the pixels within the specified threshold
    sliced_image = cv2.bitwise_and(image, image, mask=mask)
    return sliced_image

#Now, let us read the dental xray image
cap = cv2.VideoCapture("dental.gif") #since it is a gif, cv2.imread will not work. We will treat as a video.
ret, image = cap.read()
cap.release
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)#to make sure it is grayscaled



#plotting
plt.figure(figsize = (15,5))
plt.subplot(141)
plt.imshow(image, cmap = 'gray')
plt.title('original image'), plt.xticks([]), plt.yticks([])

plt.subplot(142)
plt.imshow(gray_level_slicing(image, 100,255), cmap = 'gray')
plt.title('gray level slicing at [100,255]'), plt.xticks([]), plt.yticks([])

plt.subplot(143),
plt.imshow(gray_level_slicing(image, 125,255), cmap = 'gray')
plt.title('gray level slicing at [125,255]'), plt.xticks([]), plt.yticks([])

plt.subplot(144),
plt.imshow(gray_level_slicing(image, 150,255), cmap = 'gray')
plt.title('gray level slicing at [150,255]'), plt.xticks([]), plt.yticks([])
#plt.savefig('part5a_1.png')
plt.show()

#We can also do a power law:
def power_law(image, gamma):
    power_transformed = np.power(image / 255.0, gamma) * 255.0
    power_transformed = np.uint8(power_transformed) #convert to 8-bit image
    return power_transformed


#plotting
plt.figure(figsize = (15,5))
plt.subplot(141)
plt.imshow(image, cmap = 'gray')
plt.title('original image'), plt.xticks([]), plt.yticks([])

plt.subplot(142)
plt.imshow(power_law(image, 0.5), cmap = 'gray')
plt.title('power law with gamma = 0.5'), plt.xticks([]), plt.yticks([])

plt.subplot(143),
plt.imshow(power_law(image, 1.5), cmap = 'gray')
plt.title('power law with gamma = 1.5'), plt.xticks([]), plt.yticks([])

plt.subplot(144),
plt.imshow(power_law(image, 3), cmap = 'gray')
plt.title('power law with gamma = 3.0'), plt.xticks([]), plt.yticks([])
#plt.savefig('part5a_2.png')
plt.show()



```

## (b) cells27.jpg

```python
import cv2
import numpy as np

image = cv2.imread('cells27.jpg')

#We can do a power law:
def power_law(image, gamma):
    power_transformed = np.power(image / 255.0, gamma) * 255.0
    power_transformed = np.uint8(power_transformed) #convert to 8-bit image
    return power_transformed
    

#plotting
plt.figure(figsize = (15,5))
plt.subplot(141)
plt.imshow(image, cmap = 'gray')
plt.title('original image'), plt.xticks([]), plt.yticks([])

plt.subplot(142)
plt.imshow(power_law(image, 0.5), cmap = 'gray')
plt.title('power law with gamma = 0.5'), plt.xticks([]), plt.yticks([])

plt.subplot(143),
plt.imshow(power_law(image, 0.75), cmap = 'gray')
plt.title('power law with gamma = 0.75'), plt.xticks([]), plt.yticks([])

plt.subplot(144),
plt.imshow(power_law(image, 1.5), cmap = 'gray')
plt.title('power law with gamma = 1.5'), plt.xticks([]), plt.yticks([])
#plt.savefig('part5b_1.png')
plt.show()

#We can also take the negative of the image and apply power law to it:
negative_image = 255 - image

#plotting
plt.figure(figsize = (15,5))
plt.subplot(141)
plt.imshow(negative_image, cmap = 'gray')
plt.title('negative of image'), plt.xticks([]), plt.yticks([])

plt.subplot(142)
plt.imshow(power_law(negative_image, 0.5), cmap = 'gray')
plt.title('power law with gamma = 0.5'), plt.xticks([]), plt.yticks([])

plt.subplot(143),
plt.imshow(power_law(negative_image, 1.5), cmap = 'gray')
plt.title('power law with gamma = 1.5'), plt.xticks([]), plt.yticks([])

plt.subplot(144),
plt.imshow(power_law(negative_image, 3), cmap = 'gray')
plt.title('power law with gamma = 3.0'), plt.xticks([]), plt.yticks([])
#plt.savefig('part5b_2.png')
plt.show()
```

## (c) butterfly.gif

```python
import cv2
import numpy as np

cap = cv2.VideoCapture("butterfly.gif")
ret, image = cap.read()
cap.release()

#First, we apply a median filter to remove that subtle salt and pepper noise using median blur
median_blurred = cv2.medianBlur(image, 3) #kernel  = 3
#Next, we do contrast stretching by simply normalizing the image in range 0 to 255:
contrast_stretched = cv2.normalize(median_blurred, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
#dtype=cv2.CV_8U is an 8-bit since our range is 0 to 255

#plotting
plt.figure(figsize = (10,5))
plt.subplot(131)
plt.imshow(image, cmap = 'gray')
plt.title('original image'), plt.xticks([]), plt.yticks([])

plt.subplot(132)
plt.imshow(median_blurred, cmap = 'gray')
plt.title('median blur with kernel = 3'), plt.xticks([]), plt.yticks([])

plt.subplot(133)
plt.imshow(contrast_stretched, cmap = 'gray')
plt.title('constrast stretched'), plt.xticks([]), plt.yticks([])
#plt.savefig('part5c.png')
plt.show()
```
